```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import requests
import io
```

```python
aap = pd.read_csv("AAP.csv")
ntap = pd.read_csv("NTAP.csv")
regn = pd.read_csv("REGN.csv")
sre = pd.read_csv("SRE.csv")

companies = {"AAP":aap, "NTAP":ntap, "REGN":regn, "SRE":sre}
aap.head()
```

```python
# Setting date as date time
aap["Date"] = pd.to_datetime(aap["Date"])
ntap["Date"] = pd.to_datetime(ntap["Date"])
regn["Date"] = pd.to_datetime(regn["Date"])
sre["Date"] = pd.to_datetime(sre["Date"])

print(aap.Date.dtypes)
```

```python
aap.shape
```

```python
# 1. Is there a company that has no difference between the High and Low columns? What does it mean for the stock's trading behavior on such days?
for comp, df in companies.items():
    if sum(df["High"] == df["Low"]) == df.shape[0]:
            print(f"the company {comp} has no difference between High and Low")
    else:
        print("Nope")

# There is no company that has no difference between the High and Low columns. 
```

```python
# 2. (1pt task) "What is the highest and lowest price (High) each company recorded?"
for comp, df in companies.items():
    high = df["High"].max()
    low = df["High"].min()
    print(f"Highest High price for company {comp} is {high}")
    print(f"Lowest High price for company {comp} is {low}")
```

```python
# 3. (1pt task) Calculate logarithmic returns from Close. For each company report on its, min, max, mean, standard deviation, and skewness of the return distribution.
for comp, df in companies.items():
    df["LogR_Close"] = np.log(df["Close"])-np.log(df["Close"].shift(1))
    print(f"The min logarithmic return from Close for company {comp} is {df["LogR_Close"].min()}")
    print(f"The max logarithmic return from Close for company {comp} is {df["LogR_Close"].max()}")
    print(f"The mean logarithmic return from Close for company {comp} is {df["LogR_Close"].mean()}")
    print(f"The standard deviation of logarithmic return from Close for company {comp} is {df["LogR_Close"].std()}")
    print(f"The skewness of logarithmic return from Close for company {comp} is {df["LogR_Close"].skew()}")
```

```python
# 4. When did each company record the second highest gain and second highest loss for the day? (logarithmic loss). Hint: idxmax
# Finding the first highest and lowest value
for company, df in companies.items():
    df["LogReturn"] = np.log(df["Close"])-np.log(df["Close"].shift(1))
    highest_gain_idx = df["LogReturn"].idxmax()
    highest_loss_idx = df["LogReturn"].idxmin()
    print(f"For company {company}:")
    print(f"Highest gain was on {highest_gain_idx}")
    print(f"Highest loss was on {highest_loss_idx}\n")
```

```python
# 4. When did each company record the second highest gain and second highest loss for the day? (logarithmic loss). Hint: idxmax
# Removing the higest and lowest ones from the dfs
aap_second = aap.drop(index=[219,155])
ntap_second = ntap.drop(index=[221,649])
regn_second = regn.drop(index=[1430,808])
sre_second = sre.drop(index=[361,804])

companies_red = {"aap":aap_second, "ntap":ntap_second, "regn":regn_second, "sre":sre_second}
#Repeating with the reduced data frames
for company, df in companies_red.items():
    df["LogReturn"] = np.log(df["Close"])-np.log(df["Close"].shift(1))
    highest_gain_idx = df["LogReturn"].idxmax()
    highest_loss_idx = df["LogReturn"].idxmin()
    print(f"For company {company}:")
    print(f"Second highest gain was on {df.Date[highest_gain_idx]}")
    print(f"Second highest loss was on {df.Date[highest_loss_idx]}\n")
```

```python
# 5. What is the median calendar weekly volume for each company? Hint: check how to resample pandas DF
# Set date as index
aap_date = aap.set_index("Date")
ntap_date = ntap.set_index("Date")
regn_date = regn.set_index("Date")
sre_date = sre.set_index("Date")

weekly_summary_aap = aap_date.resample('W').sum()
ans_a = weekly_summary_aap["Volume"].median()
print(f"For company aap the median calendar weekly volume is {ans_a}")

weekly_summary_ntap = ntap_date.resample('W').sum()
ans_n = weekly_summary_ntap["Volume"].median()
print(f"For company ntap the median calendar weekly volume is {ans_n}")

weekly_summary_regn = regn_date.resample('W').sum()
ans_r = weekly_summary_regn["Volume"].median()
print(f"For company regn the median calendar weekly volume is {ans_r}")

weekly_summary_sre = sre_date.resample('W').sum()
ans_s = weekly_summary_sre["Volume"].median()
print(f"For company aap the median calendar weekly volume is {ans_s}")
```

```python
# 6. Which company recorded the lowest total return over the whole period?
# total return -> (ending value/beginning value) - 1
for comp, df in companies.items():
    tot_ret = (df.Close[0]/df.Close[df.shape[0]-1])-1
    print(f"Company {comp} has the total return {tot_ret}")
```

```python
# 7. Create a new column volume_class based on the volume column into categories (e.g., "Low", "Medium", "High") and use tertile thresholds for the classes.
for comp, df in companies.items():
    tertiles = df['Volume'].quantile([1/3,2/3]).tolist()
    # Add a lower and upper range for the bins in pd.cut
    tertiles = [0] + tertiles + [float('inf')]
    # Add new tertile column to the data frame based on the tertile cut
    df['volume_class'] = pd.cut(df['Volume'], bins=tertiles, labels=['Low', 'Medium', 'High'])
```

```python
# 8. A. Plot the log-returns of the companies (ideally in the same plot).
plt.subplots(1,1, figsize = (20,7))
for company, df in companies.items():
    df["Logreturns"] = np.log(df["Close"])-np.log(df["Close"].shift(1))
    plt.plot(df["Date"],df["Logreturns"], label = company)
plt.legend()
plt.show()
```

```python
# 8. B. Show the log-return distribution of the companies (ideally in the same plot).
plt.subplots(1,1, figsize = (20,7))
for comp, df in companies.items():
    df["Logreturns"] = np.log(df["Close"])-np.log(df["Close"].shift(1))
    plt.hist(df["Logreturns"], label = comp, bins= 15, alpha = 0.4)
plt.legend()
```
